```python
import pymupdf  # import fitz, 虽然大家都用 fitz, 但为了代码提示, 还是用 pymupdf 比较好
import fitz

pdf_file_path = "./data/2301.12597v3.pdf"
pdf_bytes = open(pdf_file_path, "rb").read()
pdf_docs = pymupdf.open('pdf', pdf_bytes)
pdf_docs
```

```python
def poly_to_bbox(poly: list):
    # 需要 左上, 右下 两个点
    x0, y0, x1, y1 = poly[0], poly[1], poly[0], poly[1]
    for i in range(2, len(poly), 2):
        x0 = min(x0, poly[i])
        x1 = max(x1, poly[i])
        y0 = min(y0, poly[i + 1])
        y1 = max(y1, poly[i + 1])
    bbox = [x0, y0, x1, y1]
    bbox = [x / 2.777 for x in bbox]
    return bbox


def drop_bbox(page, bbox, i=0, rgb=None, extra_text="", fill=True):
    if rgb is None:
        rgb = [255, 0, 9]
    new_rgb = []
    for item in rgb:
        item = float(item) / 255
        new_rgb.append(item)
    rect_coords = pymupdf.Rect(*bbox)  # Define the rectangle
    if fill:
        page.draw_rect(
            rect_coords,
            color=None,
            fill=new_rgb,
            fill_opacity=0.3,
            width=0.5,
            overlay=True,
        )
    else:
        page.draw_rect(
            rect_coords,
            color=new_rgb,
            fill=None,
            fill_opacity=1,
            width=0.5,
            overlay=True,
        ) 
    x0, y0, x1, y1 = bbox
    page.insert_text(
        (x1 + 2, y0 + 10), str(i + 1) + extra_text, fontsize=10, color=new_rgb
    )  # Insert the index in the top left corner of the rectangle
```

```python
poly = [
    851.2911376953125,
    592.499267578125,
    1503.60205078125,
    592.499267578125,
    1503.60205078125,
    894.8596801757812,
    851.2911376953125,
    894.8596801757812,
]
bbox = poly_to_bbox(poly)
print(bbox)

# 保存文件
page = pdf_docs[0]
drop_bbox(page, bbox)
import os
os.makedirs("out/explore", exist_ok=True)   
pdf_docs.save("out/explore/test.pdf")
```

```python
from magic_pdf.config.ocr_content_type import CategoryId

names = [x for x in dir(CategoryId) if not x.startswith("__")]

type_dict = {}
for name in names:
    value = getattr(CategoryId, name)
    type_dict[value] = name
type_dict
```

```python
# 画出整个页面的

"""
class CategoryType(IntEnum):
     title = 0               # 标题
     plain_text = 1          # 文本
     abandon = 2             # 包括页眉页脚页码和页面注释
     figure = 3              # 图片
     figure_caption = 4      # 图片描述
     table = 5               # 表格
     table_caption = 6       # 表格描述
     table_footnote = 7      # 表格注释
     isolate_formula = 8     # 行间公式
     formula_caption = 9     # 行间公式的标号

     embedding = 13          # 行内公式
     isolated = 14           # 行间公式
     text = 15               # ocr 识别结果
"""
import json

# 每次要重新加载数据
pdf_docs = pymupdf.open('pdf', pdf_bytes)

model_list = json.loads(open(r"out\2301.12597v3\auto\2301.12597v3_model.json", "r", encoding="utf-8").read())

for model in model_list:
    layout_dets = model["layout_dets"]
    print(len(layout_dets))
    page_no = model["page_info"]["page_no"]
    for index, det in enumerate(layout_dets):
        # index 是当前页面的索引
        category_id = det["category_id"]
        if category_id not in [
            CategoryId.OcrText,
        ]:
            extra_text = " " +  type_dict[category_id]
            poly = det["poly"]
            bbox = poly_to_bbox(poly)
            page = pdf_docs[page_no]

            if category_id in [
                CategoryId.InterlineEquation_Layout,
                CategoryId.InlineEquation,
                CategoryId.InterlineEquation_YOLO,
            ]:
                rgb = [0, 255, 0]
                drop_bbox(page, bbox, index, rgb=rgb, fill=True)
            elif category_id in [
                CategoryId.TableBody,
                CategoryId.TableCaption,
                CategoryId.TableFootnote,
            ]:
                rgb = [0, 0, 255]
                drop_bbox(page, bbox, index, rgb=rgb,  extra_text=extra_text, fill=True)
            elif category_id in [
                CategoryId.ImageBody,
                CategoryId.ImageCaption,
                CategoryId.ImageFootnote,
            ]:
                rgb = [255, 255, 0]
                drop_bbox(page, bbox, index, rgb=rgb,  extra_text=extra_text, fill=True)
            else:
                drop_bbox(page, bbox, index, extra_text=extra_text, fill=False)
    
    # break

import os
os.makedirs("out/explore", exist_ok=True)   
pdf_docs.save("out/explore/test.pdf")
```
